```python
columns_to_drop = ['PoolQC','MiscFeature','Alley','Fence','MasVnrType','FireplaceQu','Id']
train = train.drop(columns=columns_to_drop)
```

```python
train.head()
```

```python
train = train.drop(index=[1299, 524])
```

```python
num_cols = train.select_dtypes(include='number').columns
train[num_cols] = train[num_cols].fillna(train[num_cols].median())

# Categorical imputation
cat_cols = train.select_dtypes(include='object').columns
for col in cat_cols:
    train[col] = train[col].fillna('None')
```

```python
skew_vals = numeric.apply(lambda x: x.skew()).sort_values(ascending=False)
print(skew_vals)
```

```python
# Threshold for high skew
threshold = 0.75
high_skew_cols = skew_vals[abs(skew_vals) > threshold].index
print(high_skew_cols)
```

```python
for col in high_skew_cols:
    train[col] = np.log1p(train[col])
```

```python
# After log transformation
numeric_feats = train.select_dtypes(include='number')
new_skew = numeric_feats.apply(lambda x: x.skew()).sort_values(ascending=False)
print(new_skew)
```
